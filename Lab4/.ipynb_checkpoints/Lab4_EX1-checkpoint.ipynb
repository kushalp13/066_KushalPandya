```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.tree import DecisionTreeClassifier, export_graphviz, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
from subprocess import call
```

```python
from sklearn import datasets, preprocessing

org_data = pd.read_csv("Dataset2.csv")
label_encoder = preprocessing.LabelEncoder()

for data in org_data:
  print(f"\n\nHeading : {data}")
  print(list(org_data[data]))
  org_data[data] = label_encoder.fit_transform(org_data[data])
  print(f"\n\nAfter the tranformation of {data}\n\n")
  print(list(org_data[data]))
```

```python
combined_features = tuple(zip(org_data["Outlook"], org_data["Temp"], org_data["Wind"], org_data["Humidity"]))

print("Outlook, Temp, Wind, Humidity\n\n")
for pair in combined_features:
    print(pair)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(combined_features, org_data["Class"], test_size = 0.1, random_state = 66)
print(X_train)
```

```python
from sklearn import metrics

dtc = DecisionTreeClassifier(criterion = "entropy")
dtc.fit(X_train, Y_train)
Y_pred = dtc.predict(X_test)
```

```python
print("Accuracy is ", metrics.accuracy_score(Y_test, Y_pred))
print(X_test)
print("y predicted : ", Y_pred)
print(f"Actual y_test {Y_test}")
```

```python
disp = metrics.plot_confusion_matrix(dtc, X_test, Y_test)
disp.figure_.suptitle("Confusion Matrix")
print(f"Confusion matrix: \n{disp.confusion_matrix}")
plt.show()
```

```python
from sklearn.tree import export_graphviz
export_graphviz(dtc, out_file='tree.dot',
               feature_names=['outlook','temperature','humidity','wind'],
               class_names=['play_no','play_yes'], 
               filled=True)

# Convert to png
from subprocess import call
call(['dot', '-Tpng', 'tree.dot', '-o', 'tree.png', '-Gdpi=600'])

# Display in python
import matplotlib.pyplot as plt
plt.figure(figsize = (15, 15))
plt.imshow(plt.imread('tree.png'))
plt.axis('off');
plt.show();
```
